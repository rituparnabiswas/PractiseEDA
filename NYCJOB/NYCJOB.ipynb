TOP 20 most popular agencies to provide jobs

```python
from collections import Counter
name_count = Counter(nycjob['AGENCY'].dropna().tolist()).most_common(20) #although null value not there but for standard coding maintained that
name_idx = [name[0] for name in name_count]
name_val = [name[1] for name in name_count]
fig,ax = plt.subplots(figsize=(18,16))
sns.barplot(x = name_idx,y=name_val ,ax =ax)
plt.title('Top ten agencies')
plt.xlabel('Agency Name')
plt.ylabel('Count')
ticks = plt.setp(ax.get_xticklabels(),rotation=90)
```

```python
from collections import Counter
name_count = Counter(nycjob['LEVEL'].dropna().tolist()).most_common() #although null value not there but for standard coding maintained that
name_idx = [name[0] for name in name_count]
name_val = [name[1] for name in name_count]
fig,ax = plt.subplots(figsize=(18,16))
sns.barplot(x = name_idx,y=name_val ,ax =ax)
plt.title('Top ten levels')
plt.xlabel('Job Level')
plt.ylabel('Count')
ticks = plt.setp(ax.get_xticklabels(),rotation=90)
```

how job time indicator and position type varies with level

```python
nycjob.groupby([nycjob['LEVEL'],nycjob['JOB_TIME_INDICATOR']])['NO_OF_POSITIONS'].sum()
```

```python
f,ax = plt.subplots(1,4,figsize=(30,20))
nycjob['JOB_TIME_INDICATOR'][nycjob['LEVEL'] == '0'].value_counts().plot.pie(explode=[0,0.2],autopct='%1.1f%%',ax=ax[0],shadow=True)
nycjob['JOB_TIME_INDICATOR'][nycjob['LEVEL'] == '1'].value_counts().plot.pie(explode=[0,0.2],autopct='%1.1f%%',ax=ax[1],shadow=True)
nycjob['JOB_TIME_INDICATOR'][nycjob['LEVEL'] == '1A'].value_counts().plot.pie(explode=[0],autopct='%1.1f%%',ax=ax[2],shadow=True)
nycjob['JOB_TIME_INDICATOR'][nycjob['LEVEL'] == '1B'].value_counts().plot.pie(explode=[0],autopct='%1.1f%%',ax=ax[3],shadow=True)

ax[0].set_title('Full time position')
```

```python
f,ax = plt.subplots(1,2,figsize=(20,7))
nycjob['LEVEL'][nycjob['JOB_TIME_INDICATOR'] == 'F'].value_counts().plot.pie(autopct='%1.1f%%',ax=ax[0],shadow=True)
nycjob['LEVEL'][nycjob['JOB_TIME_INDICATOR'] == 'P'].value_counts().plot.pie(autopct='%1.1f%%',ax=ax[1],shadow=True)

ax[0].set_title('Full time position % based on level')
ax[1].set_title('Part time position % based on level')
```

from the above plot we can say there is no part time position for level 1A,1B,3B,M1,M2,M5,4,M6,M7,M8
